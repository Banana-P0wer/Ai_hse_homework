## Постановка задачи

Задача заключалась в автоматической генерации имени функции по её телу.
В качестве данных использовались реальные функции из открытых репозиториев. В качестве модели применялась предобученная модель CodeT5+ без дообучения

Рассматривались два варианта входных данных: только исполняемый код функции и код вместе с документацией и комментариями

Качество оценивалось с помощью метрик Exact Match и ROUGE


## Итоговые результаты

|------------------------|-------------|---------|---------|---------|

| Код + комментарии      | 0.184       | 0.467   | 0.267   | 0.465   |

Добавление документации и комментариев улучшает качество предсказаний
по всем используемым метрикам.


## Основные наблюдения

Даже без дообучения CodeT5+ умеет понимать назначение функции и генерировать осмысленные имена на основе её тела

Использование документации и комментариев заметно улучшает результаты, особенно по метрикам ROUGE, которые отражают смысловое сходство имён

Exact Match является строгой метрикой

Для одной и той же функции часто существует несколько корректных имён

Во многих примерах модель предлагает альтернативные имена, которые по смыслу не хуже, а иногда и понятнее исходных

## Интерпретация ошибок

Большинство ошибок по метрике Exact Match связано не с неправильным
пониманием функции, а с формальными различиями в именах

Типичные случаи: использование синонимов, например `get` вместо `extract` и потеря части контекста, например источника данных;
упрощение имени при наличии подробной документации

Поэтому Exact Match следует рассматривать вместе с метриками ROUGE,
а не отдельно


## Ограничения подхода

Работа имеет несколько ограничений

Использовалась предобученная модель без дообучения на задаче
Длина входа ограничена 512 токенами, из-за чего длинные функции усекались
Оценка не учитывает несколько равнозначных корректных имён функций

При этом даже в таких условиях модель стабильно показывает понимание структуры и смысла кода


## Практическая значимость

Результаты показывают, что модели для анализа кода могут применяться для: генерации имён функций, подсказок при рефакторинге и анализа и документирования legacy-кода

Даже без дополнительного обучения модель способна быть полезной в качестве инструмента поддержки разработчика
